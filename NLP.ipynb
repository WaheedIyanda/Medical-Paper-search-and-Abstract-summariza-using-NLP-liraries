```python
###Importing my file 
import pandas as pd
article_data = pd.read_csv("train.csv")
article_data.head(25)
```

```python
####Data processing
import re

def clean_text(text):
    text= re.sub(r'\s+',' ', text) ### to remove extra spaces
    text= re.sub(r'[^a-zA-Z0-9\s]', '', text) ### to remove special character
    return text.lower()

#####
article_data["cleaned_title"]= article_data["title"].apply (clean_text)
article_data["cleaned_abstract"]= article_data["abstract"].apply (clean_text)
# Save cleaned data for later use
article_data.to_csv('cleaned_medical_papers.csv', index=False)
```

```python
####small querying.
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline


####### All possible Queries and label
train_queries = [
    "What are the latest treatments for diabetes?", "New therapies for heart disease",
    "Current methods for treating hypertension", "Effective treatments for arthritis",
    "How to treat migraine headaches?", "Recent advances in cancer research",
    "Latest studies on Alzheimer's disease", "New findings in immunotherapy",
    "Cutting-edge research on gene therapy", "Recent publications on vaccine development",
    "Current public health guidelines for COVID-19", "Statistics on global obesity rates",
    "Public health policies for smoking cessation", "Impact of climate change on public health",
    "Strategies for improving mental health", "New drugs approved by the FDA",
    "Side effects of common antidepressants", "Efficacy of new antiviral medications",
    "Research on drug interactions with herbal supplements", "Innovations in drug delivery systems",
    "Latest diagnostic tools for cancer detection", "Advances in imaging technologies",
    "New methods for early diagnosis of diabetes", "Techniques for genetic testing",
    "Non-invasive diagnostic procedures"
]

train_labels = [
    "treatment_query", "treatment_query", "treatment_query", "treatment_query", "treatment_query",
    "research_query", "research_query", "research_query", "research_query", "research_query",
    "public_health_query", "public_health_query", "public_health_query", "public_health_query", "public_health_query",
    "pharmacology_query", "pharmacology_query", "pharmacology_query", "pharmacology_query", "pharmacology_query",
    "diagnostic_query", "diagnostic_query", "diagnostic_query", "diagnostic_query", "diagnostic_query"
]


#### Classsification using Logistic Regresion
pipeline = make_pipeline(TfidfVectorizer(), LogisticRegression())
pipeline.fit(train_queries, train_labels)
```

```python
#### predicting intent
query = "What are the latest treatments for diabetes?"
intent = pipeline.predict([query])
print(intent)
```

```python
from elasticsearch import Elasticsearch
```

```python
es = Elasticsearch(
    ['https://localhost:9200/'],
    http_auth=('elastic', 'waheed'),
    verify_certs=False  # or provide the path to your certificate if not bypassing SSL verification
)
```

```python
for index, row in article_data.iterrows():
    es.index(index="article_data", id=row["uuid"], body=row.to_dict())

# Define search function outside of the indexing loop
def search_papers(query):
    body = {
        "query": {
            "multi_match": {
                "query": query,
                "fields": ["title^2", "abstract"]
            }
        }
    }
    res = es.search(index='article_data', body=body)
    return res['hits']['hits']
```

```python
# Search for papers
search_results = search_papers(query)
for result in search_results:
    print(result['_source']['title'])
```

```python
##### Summarization model
from transformers import pipeline

##### lading summarization model
summarizer = pipeline("summarization", model ="facebook/bart-large-cnn")

# Summarize the abstract of the top paper
top_abstract = search_results[0]['_source']['abstract']
summary = summarizer(top_abstract, max_length=150, min_length=30, do_sample=False)
print(summary[0]['summary_text'])
```
